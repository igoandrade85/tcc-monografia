```python
def secure_name(string):
    """
    Remove caracteres especiais de uma string
    """
    pattern = r'[^A-Za-z0-9_]+'
    output_string = string.replace(' ', '_').lower()
    output_string = unidecode.unidecode(output_string)
    output_string = re.sub(pattern, '', output_string)
    return output_string
```

## Indicadores de Liquidez

* Liquidez Corrente: $$LC = \dfrac{AC}{PC}$$

* Liquidez Seca: $$LS = \dfrac{AC - E - DA}{PC}$$

* Liquidez Imediata: $$LI = \dfrac{D}{PC}$$

* Liquidez Geral: $$LG = \dfrac{AC + RLP}{PC + ELP}$$

```python
def calcula_liquidez(x):
    AC = x[x['CD_CONTA']=='1.01'].iloc[0]['VL_CONTA'] # Ativo Circulante
    E = x[x['CD_CONTA']=='1.01.04'].iloc[0]['VL_CONTA'] # Estoques
    DA = x[x['CD_CONTA']=='1.01.07'].iloc[0]['VL_CONTA'] # Despesas Antecipadas
    D = x[x['CD_CONTA']=='1.01.01'].iloc[0]['VL_CONTA'] + x[x['CD_CONTA']=='1.01.02'].iloc[0]['VL_CONTA'] # Caixa e equivalente de caixa + Aplic. Financeiras
    RLP = x[x['CD_CONTA']=='1.01'].iloc[0]['VL_CONTA'] + x[x['CD_CONTA']=='1.02.01'].iloc[0]['VL_CONTA'] # Ativo Circulante + Realizável a Longo Prazo
    PC = x[x['CD_CONTA']=='2.01'].iloc[0]['VL_CONTA']
    ELP = x[x['CD_CONTA']=='2.01'].iloc[0]['VL_CONTA'] +  x[x['CD_CONTA']=='2.02'].iloc[0]['VL_CONTA']
    LC = AC/PC
    LS = (AC - E - DA) / PC
    LI = D / PC
    LG = (AC + RLP) / (PC + ELP)
    return pd.Series([LC, LS, LI, LG], index=['Corrente', 'Seca', 'Imediata', 'Geral'])
```

```python
def lista_setores():
    setores_dict = {}
    for setor_nome in novo_mercado['Setor Econômico'].drop_duplicates().to_list():
        setor_df = novo_mercado[novo_mercado['Setor Econômico']==setor_nome]\
            .sort_values(by=['Capital R$'], ascending=False)\
            .reset_index(drop=True)
        setor_df['Peso Capital Social'] = setor_df['Capital R$'] / setor_df['Capital R$'].sum()
        
        setores_dict[setor_nome] = setor_df[['Denominação Social', 'Código CVM', 'Ticker', 'Capital R$', 'Peso Capital Social']]
        
    return setores_dict
```

```python
def calcula_liquidez_setorial(dfcs, setor, nome=None):
    liquidez_empresas = {}
    erros = []
    for cd_cvm in setor['Código CVM'].to_list():
        try:
            empresa = dfcs[dfcs['CD_CVM']==cd_cvm].reset_index(drop=True)
            if nome=='Comunicações':#17639 -> Tim participações
                empresa = dfcs[dfcs['CD_CVM'].isin([cd_cvm, 17639])].reset_index(drop=True)
            empresa = empresa[empresa['CD_CONTA'].apply(lambda x: len(x) <= 10)]
            liquidez = empresa.groupby('DT_REFER').apply(calcula_liquidez)
            liquidez.index = pd.PeriodIndex(liquidez.index, freq='Q')
            liquidez_empresas[cd_cvm] = liquidez
        except:
            liquidez_empresas[cd_cvm] = pd.DataFrame()
            erros.append(cd_cvm)

    liquidez_setorial = pd.DataFrame()
    for cd_cvm, liquidez in liquidez_empresas.items():
        peso = setor[setor['Código CVM']==cd_cvm]['Peso Capital Social'].reset_index(drop=True)
        peso = peso[0]
        liquidez_setorial = liquidez_setorial.add(peso * liquidez, fill_value=0)
        
    liquidez_setorial.to_csv(os.path.join(ROOT_DIR, 'data', 'indicadores', f'liquidez_{secure_name(nome)}.csv'))

    return liquidez_setorial
```

```python
def plot_liquidez(liquidez_setorial, nome=None):
    liquidez_setorial = liquidez_setorial['2018-01-01':]
    tidy = pd.melt(liquidez_setorial.reset_index(), id_vars='DT_REFER', value_vars=['Corrente', 'Seca', 'Imediata', 'Geral'])
    tidy.columns=['Trimestre', 'Liquidez', 'Valor']

    fig, ax = plt.subplots(figsize=(20, 6))
    sns.barplot(data=tidy, x='Trimestre', y='Valor', hue='Liquidez', ax=ax)
    ax.set_ylim([0, 1.5 + int(liquidez_setorial.max().max())])

    ax.set_ylabel('Liquidez', fontweight='bold')
    ax.set_xlabel('Trimestre', fontweight='bold')
    
    plt.savefig(os.path.join(ROOT_DIR, 'img', 'indicadores', f'liquidez_{secure_name(nome)}.png'), bbox_inches='tight')

    plt.show()
```

```python
def calcula_liquidez_testes(liquidez_setorial, nome=None):
    liquidez_setorial = liquidez_setorial['2018-01-01':]
    liquidez_prepandemia = liquidez_setorial[:'2020-01-01']
    liquidez_pandemia = liquidez_setorial['2020-01-01':]

    liquidez_prepandemia.describe().drop(['count'], axis=0).to_csv(os.path.join(ROOT_DIR, 'data', 'indicadores', f'liquidez_{secure_name(nome)}_prepandemia_stats.csv'))
    liquidez_pandemia.describe().drop(['count'], axis=0).to_csv(os.path.join(ROOT_DIR, 'data', 'indicadores', f'liquidez_{secure_name(nome)}_pandemia_stats.csv'))

    cols = ['Liquidez', 'Teste', 'Estatística', 'Valor p']
    liquidez_stats = pd.DataFrame(columns=cols)
    for liq in liquidez_setorial.columns:

        group1, group2 = liquidez_prepandemia[liq].to_numpy(), liquidez_pandemia[liq].to_numpy()
        ttest= stats.ttest_ind(a=group1, b=group2, equal_var=True)
        htest = stats.kruskal(group1, group2)

        dft = pd.DataFrame([liq, 'Teste t', ttest.statistic, ttest.pvalue]).T
        dfh = pd.DataFrame([liq, 'Teste H', htest.statistic, htest.pvalue]).T

        dft.columns = cols
        dfh.columns = cols

        liquidez_stats = pd.concat([liquidez_stats, dft, dfh])

    liquidez_stats.reset_index(drop=True, inplace=True)
    liquidez_stats.to_csv(os.path.join(ROOT_DIR, 'data', 'indicadores', f'liquidez_{secure_name(nome)}_testes.csv'), index=False)
    
    return liquidez_stats
```

```python
def calcula_liquidez_geral(setores):
    liquidez_geral = pd.DataFrame(columns = ['Setor Econômico', 'Período', 'Liquidez Geral'])
    for nome in list(setores.keys()):
        for periodo in ['Pré-pandemia', 'Pandemia']:
            df = pd.read_csv(os.path.join(ROOT_DIR, 'data', 'indicadores', f'liquidez_{secure_name(nome)}_{secure_name(periodo)}_stats.csv'), index_col=0)
            lg = df.loc['mean']['Geral']
            tmp_df = pd.DataFrame([nome, periodo, lg]).T
            tmp_df.columns = ['Setor Econômico', 'Período', 'Liquidez Geral']
            liquidez_geral = pd.concat([liquidez_geral, tmp_df])

    liquidez_geral.reset_index(drop=True, inplace=True)
    return liquidez_geral
```

## Análise dos setores

```python
setores = lista_setores()

for nome in list(setores.keys()):
    print('\n', nome.upper())
    setor = setores[nome]

    print('1) Liquidez Setorial')
    liquidez_setorial = calcula_liquidez_setorial(dfcs, setor, nome=nome)
    liquidez_setorial = liquidez_setorial['2018-01-01':]
    display(liquidez_setorial)
    
    plot_liquidez(liquidez_setorial, nome=nome)
    
    print('2) Testes de Hipótese')
    
    liquidez_stats = calcula_liquidez_testes(liquidez_setorial, nome=nome)
    display(liquidez_stats)
    
print('3) Liquidez Geral')
liquidez_geral = calcula_liquidez_geral(setores)
display(liquidez_geral)
```

```python
liquidez_geral = calcula_liquidez_geral(setores)
grouped = liquidez_geral.groupby('Setor Econômico')
pandemia = grouped['Liquidez Geral'].last()
pandemia = pandemia.to_list()
lst_pan = []
for p in pandemia:
    lst_pan.append(p)
    lst_pan.append(p)

liquidez_geral['Pan'] = lst_pan
tmp = liquidez_geral.sort_values(['Pan', 'Período'], ascending=False)
fig, ax = plt.subplots(figsize=(14, 6))
sns.barplot(data=tmp, x = 'Liquidez Geral', y = 'Setor Econômico', hue='Período')
ax.axvline(1, color='red', alpha=.5, linestyle='dashed')

ax.set_xlabel('Liquidez Geral', fontweight='bold', fontsize=14)
ax.set_ylabel('Setor Econômico', fontweight='bold', fontsize=14)

ax.tick_params(axis='both', which='major', labelsize=13)
plt.legend(prop={'size': 13})

plt.savefig(os.path.join(ROOT_DIR, 'img', 'indicadores', 'liquidez_geral_setores.png'), bbox_inches='tight')
plt.show()
```
